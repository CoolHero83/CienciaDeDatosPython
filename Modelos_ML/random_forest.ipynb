# Random Forest

- Util para clasificacion y regresion 
- Utiliza varios arboles de decision durante el entrenamiento, donde realiza la prediccion basado en el resultado de la mayoria de lo arboles utilizado

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib
```

```python
datos = sns.load_dataset('iris')
datos.head()
```

```python
matriz_correlaciomn = datos.corr()
matriz_correlaciomn
```

```python
X = datos.drop('species', axis =1)
y = datos['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators = 100, random_state= 42)

model.fit(X_train, y_train)
```

```python
y_predic = model.predict(X_test)
```

```python
# mETRICAS

precicion = accuracy_score(y_test, y_predic)
print('precicion del modelo: ', precicion)
print(':::::  Reporte de clasifiacion  :::::')
print(classification_report(y_test, y_predic))
```

```python
dato_nuevo = [[0.5,0.25,1.53,1.75]]

predeccion = model.predict(dato_nuevo)
print('La prediccion es: ', predeccion)
```
