```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Creaos un conjunto de datos

datos ={
    'Gasto_publicidad':[100,200,300,400,500,100,200,300,400,500],
    'ventas' : [800,900,700,600,500,800,900,700,600,500]
}
```

```python
#Creamos un df

df = pd.DataFrame(data = datos)

#separamos las variables en predictora y objetivo

x = df[['Gasto_publicidad']]
y = df['ventas']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=42)

#modelo
model  = LinearRegression()
#entrenar modelo
model.fit(x_train, y_train)

#prediccion

y_pred = model.predict(x_test)

print('Coeficientes: ', model.coef_)
print('Pendiente: ', model.coef_[0])
print('Interseccion', model.intercept_)
```

```python
# Evaluar el modelo

from sklearn.metrics import mean_squared_error, r2_score

print("Error cuadratico medio (MSE)", mean_squared_error(y_test,y_pred))
print("Coeficiente de determinacion (R2): ", r2_score(y_test, y_pred) )
```

```python
# visualozar

plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, y_pred, color='red', linewidth=3)
plt.title('Regresion lineal: Gasto de publicidad ocntra ventas')
plt.xlabel('Gastp en publicidad')
plt.ylabel('ventas')
plt.show()
```
