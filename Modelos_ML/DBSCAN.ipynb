# BDSCAN

- Algortimo de clustering basado en la densidad que agrupa puntos de datos en regiones de alta densidad
- FUnciona encontranmdo areas en el espacio de caracteristicas deonde la densidad de los puntos es alta
- Basandose en la proximidad a ests grupos de alta densidad
- epsilon especifica la distancia maxima entre puntos para que se consideren vecinos directos


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _=make_moons(n_samples = 1000, noise=0.05)
print(X)
```

```python
# Entrenar el modelo

dbscan = DBSCAN(eps = 0.2, min_samples=5)
dbscan.fit(X)
```

```python
# visualizar los resultados del clustering

labels = dbscan.labels_
plt.scatter(X[:,0], X[:,1], c = labels, cmap = 'viridis')
plt.title('DBSCAN clustering')
plt.xlabel('Caracteristica 1')
plt.ylabel('Caracteristica 2')

plt.show()
```
